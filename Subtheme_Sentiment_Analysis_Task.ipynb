```python
def replacePosNeg(word):
    if(word.find('positive')>0):
        word = word.replace(" positive",'')
    if(word.find('negative')>0):
        word = word.replace(" negative",'')
    return word
```

```python
list_2 = []
for word in list_polar:
    list_2.append(replacePosNeg(word))
list_2 = pd.Series(list_2).unique()
list_2
```

```python
def dataModification(data,Considerpolarity):
    
    data1 = data.iloc[:, 1:]
    if(Considerpolarity==True):
        data2 = pd.DataFrame(index= range(0, data1.shape[0]), columns=list_polar)
    else:
        data2 = pd.DataFrame(index= range(0, data1.shape[0]), columns=list_2)
    
    for i, row in data1.iterrows():
        
        l_subthemes = row.dropna().tolist()
        
        for t in l_subthemes:
            
            if(Considerpolarity==True):
                data2.loc[i][t] = 1
            else :
                data2.loc[i][replacePosNeg(t)] = -1
                

    data2 = data2.fillna(0)
    finalData = pd.DataFrame()
    finalData['Sentence'] = data['Sentence']
    for column in data2.columns:
        finalData[column] = data2[column]
    return finalData
```

```python
finalData = dataModification(data,True)
finalData.head(20)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import SpatialDropout1D
```

```python
sent_length=25
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
data_pred = finalData.iloc[:,1:]
data_pred.head()
```

```python
X = np.array(embedded_docs)
Y = np.array(data_pred)
```

```python
EMBEDDING_DIM = 100 
model = Sequential()
model.add(Embedding(voc_size, EMBEDDING_DIM, input_length=sent_length))
model.add(LSTM(250, dropout=0.2,return_sequences=True))
model.add(LSTM(150, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(33, activation='sigmoid'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=5,batch_size=64)
```

```python
y_pred=np.argmax(model.predict(X_test), axis=-1)
len(y_pred)
```
